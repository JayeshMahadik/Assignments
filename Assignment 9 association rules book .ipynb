```python
#Import the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Import the dataset
book = pd.read_csv('book.csv')
book
```

APRIOI ALGORITHM

```python
#Association with taking in consideration 10% support and 70% confidence
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
#With 70% confidence
rules = association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
#Chcking lift > 1 
rules[rules.lift>1]
```

```python
#Visualization of data
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

Trying Same Association rules with 20% support and 80% Confidence

```python
frequent_itemsets2=apriori(book,min_support=0.2,use_colnames=True)
frequent_itemsets2
```

```python
#With 80% confidence
rules2 = association_rules(frequent_itemsets2,metric='lift',min_threshold=0.8)
rules2
```
