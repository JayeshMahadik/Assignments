```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
#Importing the libraries 
```

```python
#importing the dataset 
wines=pd.read_csv("wine.csv")
wines
```

```python
#describing the data 
wines.describe()
```

```python
wines.info()
```

```python
wines2=wines.iloc[:,1:]
wines2
```

```python
wines2.shape
```

```python
#converting into array form 
wines_array=wines2.values
wines_array
```

```python
win_normal=scale(wines_array)
win_normal
#Normalising the data 
```

```python
#PCA Initialising 
#Applying PCA fit transform to dataset
pca=PCA(n_components=13)
wine_pca=pca.fit_transform(win_normal)
wine_pca
```

```python
#Display PCA Co variance Matrix
pca.components_
```

```python
#The amount of variance each PCA holds 
var = pca.explained_variance_ratio_
var
```

```python
#calculating Cumulative Sum variance for each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
#Plotting variance plot for PCA components obtained 
plt.plot(var1,color='red')
```

```python
#displaying First PC values 
wine_pca[:,0:1]
```

```python
#Final dataframe adding PCA values
final_df=pd.concat([wines['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
#visualization of PCA from final dataframe obtained 
import seaborn as sns 
fig=plt.figure(figsize=(18,14))
sns.scatterplot(data=final_df)
```

Cross Checking with Other clustering Methods 

```python
#1]HIERARCHICAL CLUSTERING 
#Importing the recquired libraries 
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
#As we already have standardised version of data we directly to the dendogram plot
plt.figure(figsize=(12,10))
dendrogram = sch.dendrogram(sch.linkage(win_normal,method='complete'))
```

```python
#Creating Clusters 
hcluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hcluster
```

```python
#save clusters for chart
y_hc = pd.DataFrame(hcluster.fit_predict(win_normal),columns=['clustersid'])
y_hc['clustersid'].value_counts()
```

```python
#Adding Clusters to the Dataset 
wines3 = wines.copy()
wines3['clustersid']=hcluster.labels_
wines3
```

```python
#2]K-MEANS CLUSTERING 
from sklearn.cluster import KMeans
```

```python
#Again as we already have an normalized data we polt the elbow graph
#How to find optimum number of clusters 
#The k-Means algorithm tends to choose centroids that minimise the inertia or within cluster sum of squares criterion i.e.WCSS
```

```python
wcss = []
for i in range(1,8):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(win_normal)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,8),wcss)  
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Therefore build cluster algorithm using k = 3 
clusterkm=KMeans(3,random_state=12)
clusterkm.fit(win_normal)
clusterkm
```

```python
clusterkm.labels_
```

```python
#Adding clusters obtained by kmeans to dataset 
wines4=wines.copy()
wines4['clusterkmid']=clusterkm.labels_
wines4
```

```python
wines4['clusterkmid'].value_counts()
```
