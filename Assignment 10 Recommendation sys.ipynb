```python
#Importing the basic libraries
import pandas as pd
import numpy as np
```

```python
#Importing the dataset and displaying
books=pd.read_csv("book.csv",encoding='Latin1')
books
```

```python
#As we observed there are extra index columns in our dataset hence removing 
books2=books.iloc[:,1:]
books2
```

```python
#Sorting of dataset based on userid column
books2.sort_values('User.ID')
```

```python
#checking the number of unique users in dataset
len(books2['User.ID'].unique())
```

```python
#Displaying userid numbers based on ratings count
books2['Book.Rating'].value_counts()
```

```python
#checking number of unique books in dataset 
len(books2['Book.Title'].unique())
```

```python
#displaying book title based on value counts 
books2['Book.Title'].value_counts()
```

```python
#coverting data in pivot dataset form using pivot method
books_pv=books2.pivot_table(index='User.ID',
                      columns='Book.Title',
                      values='Book.Rating').reset_index(drop=True)
books_pv
```

```python
#Substituting Index with user id unique values 
books_pv.index=books2['User.ID'].unique()
books_pv
```

```python
#Impute those NAN with 0 values 
books_pv.fillna(0,inplace=True)
books_pv
```

```python
#Calculating cosine similarity between users in dataset
#First importing the recquired libraries for executing cosine
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1-pairwise_distances(books_pv.values,metric='cosine')
user_sim
```

```python
#Storing the results in dataframe format 
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
#Setting the user names and columns to user id 
user_sim_df.index=books2['User.ID'].unique()
user_sim_df.columns=books2['User.ID'].unique()
user_sim_df
```

```python
#Nullifying diagonal values 
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
#checking most similar users 
user_sim_df.idxmax(axis=1)
```
